```python
import os
import gensim
import numpy as np
import pandas as pd
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
```

```python
#Dataset : https://www.kaggle.com/datasets/khulasasndh/game-of-thrones-books
```

```python
nltk.download('punkt')
```

```python
story=[]
for filename in os.listdir('/content/data'):
  f=open(os.path.join('/content/data',filename),encoding="cp1252")
  corpus=f.read()
  raw_sent=sent_tokenize(corpus)
  for sentence in raw_sent:
    story.append(simple_preprocess(sentence))
```

```python
story
```

```python
model=gensim.models.Word2Vec(
    window=10,
min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('john')
```

```python
model.wv.most_similar('targaryen')
```

```python
model.wv.doesnt_match(['jon','rikon','jaime','arya','sansa','bran'])
```
